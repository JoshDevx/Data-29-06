```python
import pandas as pd

ruta = '/content/drive/MyDrive/Libro1.csv'
'''
df = pd.read_csv(ruta, sep=';', engine='python', on_bad_lines='skip')
df.head()
'''
# Intenta con estos parámetros:
df = pd.read_csv(ruta,
                sep=';',
                encoding='utf-8',  # o 'latin1' si hay caracteres especiales
                on_bad_lines='warn',  # Para ver líneas problemáticas
                dtype=str)  # Lee todo como texto inicialmente
```

```python
# Muestra las columnas cargadas
print(df.columns.tolist())

# Muestra información del DataFrame
df.info()

# Muestra datos de ejemplo
df.head(3)
```

```python
# Paso 1: Corregir nombres de columnas
df = df.rename(columns={'AÐO': 'AÑO'})

# Paso 2: Convertir columnas numéricas
columnas_numericas = ['FOB', 'CIF', 'PESO_NETO', 'ADV', 'SAD', 'IGV', 'ISC', 'IPM', 'DES', 'DER', 'CANTIDAD']
for col in columnas_numericas:
    df[col] = pd.to_numeric(df[col].str.replace(',', '.'), errors='coerce')

# Paso 3: Convertir fechas
df['FECH_INGSI'] = pd.to_datetime(df['FECH_INGSI'], format='%Y%m%d')
df['Día_final'] = pd.to_numeric(df['Día_final'], errors='coerce')
df['AÑO_final'] = pd.to_numeric(df['AÑO_final'], errors='coerce')
df['MES_final'] = pd.to_numeric(df['MES_final'], errors='coerce')

# Paso 4: Verificar resultados
print(df.dtypes)
df.head(3)
```

```python
# Verificar valores nulos
print(df.isnull().sum())

# Estadísticas descriptivas
print(df.describe(include='all'))
```

```python
# Convertir columnas categóricas
columnas_categoricas = ['VIA', 'PAIS_ORIGE', 'UNIDAD', 'ZONA', 'CONTINENTE', 'GRUPO', 'GRUPO_1', 'GRUPO_2']
for col in columnas_categoricas:
    df[col] = df[col].astype('category')
```

```python
df.to_csv('/content/drive/MyDrive/Libro1_limpio.csv', index=False, encoding='utf-8-sig')
```

```python
# Agrupar por país y calcular métricas
resumen_paises = df.groupby('PAIS_ORIGE').agg({
    'FOB': 'sum',               # Suma del valor FOB
    'CANTIDAD': 'sum',          # Suma de la cantidad importada
    'PESO_NETO': 'sum',         # Suma del peso total
    'DNOMBRE': 'count'          # Número de operaciones
}).rename(columns={'DNOMBRE': 'NÚMERO_IMPORTACIONES'})

# Ordenar por valor FOB (de mayor a menor)
top10_fob = resumen_paises.sort_values('FOB', ascending=False).head(10)

# Mostrar resultados
print("TOP 10 PAÍSES POR VALOR FOB (MONTO IMPORTADO)")
print(top10_fob)
```

```python
# Agrupar por continente
resumen_continente = df.groupby('CONTINENTE').agg({
    'FOB': 'sum',
    'CANTIDAD': 'sum',
    'PESO_NETO': 'sum'
})

# Gráfico de torta (distribución del FOB por continente)
plt.figure(figsize=(8, 8))
resumen_continente['FOB'].plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    labels=None,
    shadow=True
)
plt.title('Distribución del Valor FOB por Continente', fontsize=14)
plt.legend(labels=resumen_continente.index, bbox_to_anchor=(1, 1))
plt.ylabel('')
plt.show()
```
